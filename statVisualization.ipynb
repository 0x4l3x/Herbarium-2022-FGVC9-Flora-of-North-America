```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

# ResNet50

```python
with open('/home/amorante/logs50.pkl', 'rb') as file:
          logs = pickle.load(file)
```

```python
tloss, vloss, acc= [], [], []
i=0
for dic in logs:
    if i%2==0:
        tloss.append(list(dic.values())[0])
    else:
        vloss.append(list(dic.values())[0])
        acc.append(list(dic.values())[1])
    i+=1
i
```

```python
fig = plt.figure(figsize=(25,12))
ax = fig.add_subplot()

ax.plot(range(20), tloss, 'r--o', label="train_loss")
ax.plot(range(20), vloss, 'g--o', label="validation_loss")
ax.set_xlabel("Epoch")
ax.set_ylabel("Loss")
ax.set_title("Loss evolution per epoch")
ax.legend()
plt.xticks(range(20))
plt.show()
```

```python
fig = plt.figure(figsize=(25,12))

ax = fig.add_subplot()
ax.plot(range(20),acc,  label="Accuracy")

ax.set_xlabel("Epoch")
ax.set_ylabel("Accuracy")
ax.set_title("Accuracy evolution per epoch")
ax.legend()
plt.xticks(range(20))

plt.show()
```

```python
with open('/home/amorante/trainloss50.pkl', 'rb') as file:
          trainloss = pickle.load(file)
```

```python
num_epochs=20
num_steps=13120
stepbyepoch=num_steps//num_epochs
steps=range(0, num_steps+1, stepbyepoch)
epochs=range(num_epochs+1)
labs=[]
for i in epochs:
  labs+=["{}\n{}".format(steps[i], epochs[i])]
```

```python
fig = plt.figure(figsize=(25,12))
ax = fig.add_subplot()


ax.set_xticks(range(0, num_steps+1, stepbyepoch))
ax.set_xticklabels(labs)

joined=[]
for x in trainloss:
  joined+=x

plt.plot(joined, 'r-+')
plt.title("Train loss per setep and epoch")

plt.xlabel("Step\nEpoch")
plt.ylabel("Train loss")
plt.grid()

plt.show()
```

```python
fig = plt.figure(figsize=(50,25))
ax = fig.add_subplot(222)
ax.set_prop_cycle('color',plt.cm.hsv(np.linspace(0,1,30)))
for x in range(num_epochs):
  ax.plot(trainloss[x], '-+', label=x)
plt.title("train_batch_losses")

plt.xlabel("Train loss evolution per batch")
plt.ylabel("Steps")
plt.grid()
plt.legend(loc='best')

plt.show()
```
